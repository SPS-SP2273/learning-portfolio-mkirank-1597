**Exercise 1**

```python
from matplotlib import pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker="o", markerfacecolor='yellow',
             ecolor="black",capsize=5,
             label='Y value with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend() #FIXED
plt.show()
```

**Exercise 2**

```python
import matplotlib as mp
from matplotlib import pyplot as plt
import numpy as np

x = list(np.arange(0, 51, 0.1))
y = list()
for i in x:
    y.append((np.exp(-i/10))*(np.sin(i)))

plt.plot(x,y)
plt.ylim(-1,1)
plt.xlabel('x', color='darkorange', fontsize=15)
plt.ylabel('y', color='darkorange', fontsize=15)
plt.title("y = $\\exp^{-x/10} \\sin(x)$", fontsize=20)
plt.grid(alpha=.25)
plt.show()
```
